### Categorical features
Generate Box plots for the different feature that hold categorical values. These features would be "Category", "GPU", "OS", "CPU_core", "RAM_GB", "Storage_GB_SSD"

```python
sns.boxplot(x= "Category", y = "Price", data = df)
```

```python
# List of features to plot
features = ['Category', 'GPU', 'OS', 'CPU_core', 'RAM_GB', 'Storage_GB_SSD']

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 10))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plot each feature in a separate subplot
for i, feature in enumerate(features):
    sns.boxplot(x=feature, y='Price', data=df, ax=axes[i])
    axes[i].set_title(f'Boxplot of Price by {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Price')

# Adjust layout
plt.tight_layout()
plt.ylim(0,)

# Show the plot
plt.show()
```

```python
# List of features to plot
features = ['Category', 'GPU', 'OS', 'CPU_core', 'RAM_GB', 'Storage_GB_SSD']

# Define a palette for each feature
palettes = {
    'Category': 'Set2',
    'GPU': 'Set3',
    'OS': 'Paired',
    'CPU_core': 'Pastel1',
    'RAM_GB': 'Accent',
    'Storage_GB_SSD': 'Dark2'
}

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 10))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plot each feature in a separate subplot
for i, feature in enumerate(features):
    sns.boxplot(x=feature, y='Price', data=df, ax=axes[i], palette=palettes[feature])
    axes[i].set_title(f'Boxplot of Price by {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Price')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

### Task 3 - GroupBy and Pivot Tables
Group the parameters "GPU", "CPU_core" and "Price" to make a pivot table and visualize this connection using the pcolor plot.

```python
# Create the group
df_gptest = df[["GPU", "CPU_core", "Price"]]
df_gptest.head(4)
```

```python
grouped_test1 = df_gptest.groupby(['GPU','CPU_core'],as_index=False).mean()
print(grouped_test1)
```

```python
# Create the Pivot table
grouped_pivot = grouped_test1.pivot(index='GPU',columns='CPU_core')
print(grouped_pivot)
```

```python
#Create the Plot
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

#label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

fig.colorbar(im)
#plt.show()
```

#### Task 4 - Pearson Correlation and p-valuesÂ¶
Use the scipy.stats.pearsonr() function to evaluate the Pearson Coefficient and the p-values for each parameter tested above. This will help you determine the parameters most likely to have a strong effect on the price of the laptops.

```python
for param in ['RAM_GB','CPU_frequency','Storage_GB_SSD','Screen_Size_inch','Weight_pounds','CPU_core','OS','GPU','Category']:
    pearson_coef, p_value = stats.pearsonr(df[param], df['Price'])
    print(param)
    print("The Pearson Correlation Coefficient for ",param," is", pearson_coef, " with a P-value of P =", p_value)
```
