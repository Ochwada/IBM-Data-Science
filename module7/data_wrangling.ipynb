```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
headers = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration", 
           "num-of-doors", "body-style","drive-wheels", "engine-location", 
           "wheel-base", "length", "width", "height", "curb-weight", "engine-type",
           "num-of-cylinders", "engine-size", "fuel-system", "bore", "stroke", "compression-ratio", 
           "horsepower","peak-rpm", "city-mpg", "highway-mpg", "price"]

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv(url, header= None, names=headers)
df.head(5)
```

```python
df.describe(include="all")
```

```python
df.replace("?", np.nan, inplace= True) # replace ? with NaN
df.head()
```

```python
missing_data = df.isnull()
missing_data.head()
```

```python
# Count missing values in each column

for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print(" ")
```

```python
# Calculate the mean value for the "normalized-losses" column
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
# Replace "NaN" with mean value in "normalized-losses" column
df["normalized-losses"] = df["normalized-losses"].replace(np.nan, avg_norm_loss)
```

```python
# Calculate the mean value for the "bore" column
avg_bore = df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
#Replace "NaN" with the mean value in the "bore" column
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

### Question #1: 
Based on the example above, replace NaN in "stroke" column with the mean value.

```python
mean_stroke = df['stroke'].astype("float").mean(axis=0)
print("Mean of Stroke:", mean_stroke)
```

```python
df["stroke"].replace(np.nan, mean_stroke, inplace=True)
```

```python
#Calculate the mean value for the "horsepower" column
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
```

```python
# Replace "NaN" with the mean value in the "horsepower" column
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
#Calculate the mean value for "peak-rpm" column

avg_peakrpm = df['peak-rpm'].astype('float').mean(axis=0)
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
# To see which values are present in a particular column, we can use the ".value_counts()" method
df["num-of-doors"].value_counts()
```

```python
# You can see that four doors is the most common type. We can also use the ".idxmax()" 
# #method to calculate the most common type automatically:

df["num-of-doors"].value_counts().idxmax()
```

```python
# The replacement procedure is very similar to what you have seen previously:
df["num-of-doors"].replace(np.nan, "four",inplace= True)
```

```python
# Finally, drop all rows that do not have price data:
# simply drop whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace= True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)

df.head()
```

```python
df.dtypes
```

```python
# Convert data types to proper format
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["price"]] = df[["price"]].astype("float")
```

```python
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
```

```python
df.dtypes
```

```python
df.head()
```

```python
# Convert mpg to L/100km by mathematical operation (235 divided by mpg)
df['city-L/100km'] = 235/df["city-mpg"]

# check your transformed data
df.head()
```

```python
df["highway-L/100km"] = 235/df["highway-mpg"]
df.head()
```

```python
# replace (original value) by (original value)/(maximum value)

df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
```

```python
df['height'] = df['height']/df['height'].max()

# show the scaled columns
df[["length", "width","height"]].head()
```

```python
#Binning
# Convert data to correct format:
df["horsepower"] = df["horsepower"].astype(int, copy= True)
```

```python
# Plot the histogram of horsepower to see the distribution of horsepower.

%matplotlib inline
import matplotlib.pyplot as plt

plt.hist(df["horsepower"])

# set x/y labels and plot title
plt.xlabel("Horsepower")
plt.ylabel("Count")
plt.title("Histogram of Horsepower")
```

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
# Set group names:
group_names = ["low", "medium", "High"]
```

```python
#Apply the function "cut" to determine what each value of df['horsepower'] belongs to.
df["horsepower-binned"] = pd.cut(df["horsepower"], bins, labels=group_names, include_lowest= True)
df[["horsepower", "horsepower-binned"]].head(10)
```

```python
# See the number of vehicles in each bin:
df["horsepower-binned"].value_counts()
```

```python
plt.hist(df["horsepower-binned"], bins =3)

plt.xlabel("horsepower-binned")
plt.ylabel("count")
plt.title("horsepower bins")
```

```python
import matplotlib as plt
from matplotlib import pyplot

pyplot.bar(group_names, df["horsepower-binned"].value_counts())
# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
# Get the indicator variables and assign it to data frame "dummy_variable_1":
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```

```python
##Change the column names for clarity:
dummy_variable_1.rename(
    columns={'gas': 'fuel-type-gas', 'diesel': 'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

```python

# In the data frame, column 'fuel-type' now has values for 'gas' and 'diesel' as 0s and 1s.
# merge data frame "df" and "dummy_variable_1"

df = pd.concat([df, dummy_variable_1], axis = 1)

# drop original column "fuel-type" from "df"
df.drop("fuel-type", axis=1, inplace=True)

df.head()
```

```python
#Similar to before, create an indicator variable for the column "aspiration"

dummy_variable_2 = pd.get_dummies(df["aspiration"])
dummy_variable_2.head()
```

```python
# change column names for clarity
dummy_variable_2.rename(
    columns={'std': 'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=True)

# show first 5 instances of data frame "dummy_variable_1"
dummy_variable_2.head()
```

```python
# merge data frame "df" and "dummy_variable_2"
df = pd.concat([df, dummy_variable_2], axis=1)

#drop original column "aspiration"
df.drop("aspiration", axis= 1, inplace= True)
df.head()
```
