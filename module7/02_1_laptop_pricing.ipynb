```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filepath = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_mod1.csv"
df = pd.read_csv(filepath)
df.head()
```

```python
df.info()
```

```python

# Note that we can update the Screen_Size_cm column such that all values are rounded to nearest 2 decimal places by using numpy.round()
df["Screen_Size_cm"] = np.round(df[["Screen_Size_cm"]], 2)
df.head(5)
```

```python
df.describe(include= "all")
```

```python
missing_values = df.isnull()
missing_values.head()
```

```python
for column in missing_values.columns.values.tolist():
    print(column)
    print(missing_values[column].value_counts())
    print("")
```

```python
# Replace the missing values of weight with the average value of the attribute.
mean_weight_kg = df["Weight_kg"].astype('float').mean(axis=0)
#print(mean_weight_kg)
df["Weight_kg"].replace(np.nan, mean_weight_kg, inplace=True)
df["Weight_kg"].tail()
```

```python
# change data m type
df["Weight_kg"] = df["Weight_kg"].astype("float")
df["Screen_Size_cm"] = df["Screen_Size_cm"].astype("float")

# df[["Weight_kg","Screen_Size_cm"]] = df[["Weight_kg","Screen_Size_cm"]].astype("float")

df.dtypes
```

```python
df.head()
```

```python

# Data Standardization
# transform cm to inch by mathematical operation (1 inch = 2.54 cm)

df["Screen_Size_cm"] = df["Screen_Size_cm"]/ 2.54
df["Weight_kg"] = df["Weight_kg"]*2.205

# rename column name from "Screen_Size_cm" to "Screen_Size_inch" (also weights)
df.rename(columns={"Screen_Size_cm": "Screen_Size_inch", "Weight_kg": "Weight_pounds"}, inplace= True)


# check your transformed data 
df.head()
```

```python
df["CPU_frequency"]= df["CPU_frequency"]/ df["CPU_frequency"].max()
```
